# CORGIS Graduates Dataset 

The dataset can be accessed [here.](https://corgis-edu.github.io/corgis/python/graduates/)
Data collected from the National Survey of Recent College Graduates. 
Data includes information about employment numbers, major information, and salary. 


```python
import graduates
```

```python
grad_major = graduates.get_grad_major()
grad_major
```

The core analysis of this data is based off of the salaries of each major. In this preliminary exploration of the data, I wanted to gather data about the mean of salary of each major. In the future, more data can be explored from each major, as well as more meticulous analysis using data besides the mean. 

Initially, I separated the majors and mean salary data from the dataset. 

```python
#create a new dictionary that will contain info on each major and their mean salary 
majorsAndSal = {}
for val in grad_major: 
    #extract the name of the major and mean salary from the dataset 
    majorsAndSal.update( {(val['Education'])['Major'] : (val['Salaries'])['Mean']} )

#this should now contain dictionary of majors and their mean salary 
#majorsAndSal
```

A helpful figure to have is always the mean.

```python
#creates a function that returns the mean of the values in a dictionary 
def getMean(d): 
    sum = 0
    for key in d: 
        sum += d[key] 
    return sum/ len(d) 
```

Using this function on our data gives us the mean of the mean of salaries of all majors, in other words the mean salary of all students regardless of major. 

```python
meanSalary = getMean(majorsAndSal)
```

Function that returns the "high" salaries of majors, "high" being defined as above the mean. 

```python

def high (d):
    newDictHigh = {} 
    #newDictLow = {}
    for key in d: 
        if d[key] > meanSalary: 
            newDictHigh.update({key : d[key]})
     #   else:
      #      newDictLow.update({key, d[key]})
    
    
    return newDictHigh

highSalaries = high(majorsAndSal)
highSalaries             

    
        
```

Function that returns the "low" salaries of majors, "low" being defined as below the mean. 

```python

def low (d): 
    newDictLow = {} 
   
    for key in d: 
        if d[key] < meanSalary: 
            newDictLow.update({key : d[key]})
     
    
    return newDictLow

lowSalaries = low(majorsAndSal)
lowSalaries
```
